complex model:
1) Building envelop: 3R1C -> one fast response path (R) and one slow response path (R-C-R)
2) inside each zone: 2C1R -> indoor air (C) and building fabric walls, floors (C)
3) between zone: 3R1C -> one fast response path (R) and one slow response path (R-C-R)

simplfied model:
1) 1R
2) 1R1C
3) 1R1C

C*T_dot = u + d + (T_out -T)/R

C*(T(k+1) - T(k))/delta_t = u(k) + d(k) + (T_out(k) - T(k))/R

Input: T_out, u, d
Output: T
Parameter: R, C, 

Potential modelling (consider only 1 inside thermal zone per floor):
1) 3R2C
lumped R and lumped C, 1R for between zone, two floors

2) 5R2C
1R for envelop; 1R1C for inside zone, 1R for between zone; two floors
T_out, T_2f_ex, T_2f_in, T_1f_ex, T_1f_ex, u, d

3) 11R7C
3R1C for envelop; 1R2C for inside zone, 3R1C for between zones; two floors

```python
from __future__ import print_function
from random import random
from random import seed
seed(1)
import math

%matplotlib inline
# importing the required module 
import matplotlib.pyplot as plt 
plt.rcParams['figure.figsize'] = [10, 8] # for square canvas
from copy import copy, deepcopy
from tqdm import tqdm
```

```python
# Solar information
f2w_ratio = 0.15
shade_C = 0.75
A = 104000/4*f2w_ratio
SHGC = 0.4
```

```python
meanAFerr = []
aFerrmeanavgn = [] 
n = []
ntemp = []
gen = []
n1avg = []
n2avg = []
n3avg = []
n4avg = []
n5avg = []

ND = 10 # number of data vectors in array
DI = 5  # number of data items in vectors
NS = 10 # number of DNA strands

NGEN = 500        # number of generations (steps)
MFRAC = 0.75      # faction of median threshold
D = 2            # number of days training

'''guesses for initial solution population'''
n1i = 10    # this is R1
n2i = 10    # this is R2
n3i = 10    # this is R3
n4i = 50    # this is C1
n5i = 50    # this is C2

# here the number of data vectors equals the number of DNA strands (or organisms)
# they can be different if they are randomly paired to compute Ferr (survivability) function
for k in range(NGEN):
    gen.append(k)   # generation array stores the
    meanAFerr.append(0.0)
    aFerrmeanavgn.append(0.0)
    n1avg.append(0.0)
    n2avg.append(0.0)
    n3avg.append(0.0)
    n4avg.append(0.0)
    n5avg.append(0.0)
```

```python
#- initialize arrays before start of evolution loop EL
#then - create array of DNA strands n[i] and ntemp[i] with dimesnion NS = 5

#i initialize array where rows are dna vectors [n0i,n1i,...n5i] with random pertubations

for i in range(ND):
    n.append([n1i+0.1*random(), n2i+0.1*random(), n3i+0.1*random(), n4i+0.1*random(), n5i+0.1*random()])
#print (n) # uncomment command to print array so it can be checked

ntemp = deepcopy(n)
#initialize Ferr values an dother loop parameters
#define arrays of Ferr (error) functions
#individual solution error and absoute error
Ferr = []  
#population average solution error and absoute error
Ferravgn = []
aFerr = []
aFerravgn = []

#store zeros in ND genes
for i in range(ND):
    #individual solution error and absoute error
    Ferr.append([0.0])
    aFerr.append([0.0])
    #population average solution error and absoute error
    Ferravgn.append([0.0])
    aFerravgn.append([0.0])
#print (Ferr)

aFerrmeanavgnMin = 1000000000.0 
# these store the  n values for minimum population average error durng NGEN generations
n1min = 0.0
n2min = 0.0
n3min = 0.0
n4min = 0.0
n5min = 0.0

aFerrta = 0.0
# these store the time averaged n values durng from generation 800 to NGEN generations
n1ta = 0.0
n2ta = 0.0
n3ta = 0.0
n4ta = 0.0
n5ta = 0.0
```

```python
'''START OF EVOLUTION LOOP'''
# ------------------------------------------
# k is generation number, NGEN IS TOTAL NUMBER OF GENERATIONS COMPUTED
for k in tqdm(range(NGEN)):
    
    '''In this program , the number of organisms (solutions) NS is taken to be equal to the 
    number of data points ND so for each generation, each solution can be compared to a different 
    data point and all the data is compared in each generation.  The order of the solutions in the array 
    that holds the solution constants is constantly changing due to mating and selection, so the pairing 
    is random.'''
    
    
    '''CALCULATING ERROR (FITNESS)
    In this program, the absolute error in the logrithm of the physical heat flux is 
    used to evaluate fitness.'''
    
    # Here we calculate error Ferr and absolute error aFerr for each data point and organism pair 
    # for specified n(i), and calculate (mean aFerr) = aFerrmean
    # and (median aFerr) = aFerrmedian for the data collection and specified n(i)
    # Note that the number data points ND equals the number of solutions (organisms) in the population
    #====================================================================================
    for i in range(ND):
        error = 0
        for j in range(int(D*24*60/30)):
            zoneT_NF_avg = np.mean([np.mean(zoneT_in_NF_data_df.iloc[j, 1:-4]), np.mean(zoneT_ex_NF_data_df.iloc[j, 1:-4])])
            zoneT_NS_avg = np.mean([np.mean(zoneT_in_NS_data_df.iloc[j, 1:-4]), np.mean(zoneT_ex_NS_data_df.iloc[j, 1:-4])])
            zoneT_NF_avg_next = np.mean([np.mean(zoneT_in_NF_data_df.iloc[j+1, 1:-4]), np.mean(zoneT_ex_NF_data_df.iloc[j+1, 1:-4])])
            zoneT_NS_avg_next = np.mean([np.mean(zoneT_in_NS_data_df.iloc[j+1, 1:-4]), np.mean(zoneT_ex_NS_data_df.iloc[j+1, 1:-4])])
            outdoorT = np.mean([outdoor_data_df['air_temp_set_1'][j], outdoor_data_df['air_temp_set_2'][j]])
            T_Sf_calc = ((outdoorT - zoneT_NS_avg)/(n[i][0]*n[i][3]) + 
                         (zoneT_NF_avg - zoneT_NS_avg)/(n[i][1]*n[i][3]) + 
                         (energy_data_df['hvac_N'][j]/2)/n[i][3] + 
                         (energy_data_df['mels_N'][j]/2)/n[i][3] +
                         (outdoor_data_df['solar_radiation_set_1'][j]/1000*0.5*A*SHGC*(1 - shade_C)/n[i][3])*0.5) + zoneT_NS_avg
            T_Ff_calc = ((outdoorT - zoneT_NF_avg)/(n[i][2]*n[i][4]) + 
                         (zoneT_NS_avg - zoneT_NF_avg)/(n[i][1]*n[i][4]) + 
                         (energy_data_df['hvac_N'][j]/2)/n[i][4] +
                         (energy_data_df['mels_N'][j]/2)/n[i][4] +
                         (outdoor_data_df['solar_radiation_set_1'][j]/1000*0.5*A*SHGC*(1 - shade_C)/n[i][4])*0.5) + zoneT_NF_avg
            error = error + np.mean([np.abs(T_Ff_calc - zoneT_NF_avg_next), np.abs(T_Sf_calc - zoneT_NS_avg_next)])

        aFerr[i] = error/(D*24*60/30) # Assign error for book keeping
    #-------------
    aFerrmean = np.mean(aFerr) # mean error for population for this generation
    aFerrmedian = np.median(aFerr) # median error for population for this generation
    meanAFerr[k]=aFerrmean  #store aFerrmean for this generation gen[k]=k


    '''SELECTION'''
    #pick survivors 
    #[2] calculate survival cutoff, set number kept = nkeep = 0
    #==============================================
    clim = MFRAC*aFerrmedian  #cut off limit is a fraction/multiplier MFRAC of median value for population
    nkeep = 0
    
    # now check each organism/solution to see if aFerr is less than cut of limit
    #if yes, store n for next generation population in ntemp, at end nkeep = number that survive
    #and number of new offspring = NS-nkeep
    #==============================================
    for j in range(NS):  # NS Ferr values, one for each solution in population
        if (aFerr[j] < clim):
            nkeep = nkeep + 1
            ntemp[nkeep-1][0] = n[j][0]
            ntemp[nkeep-1][1] = n[j][1]
            ntemp[nkeep-1][2] = n[j][2]
            ntemp[nkeep-1][3] = n[j][3]
            ntemp[nkeep-1][4] = n[j][4]

    #now have survivors in leading entries in list of ntemp vectors from 1 to nkeep
    #compute number to be added by mating
    nnew = NS - nkeep  
    
    '''MATING'''
    mut_c = 0.1
    #[4] for nnew new organisms/solutions,
    # randomly pick two survivors, randomly pick DNA (n) from pair for each offspring
    #==============================================================
    for j in range(nnew): 
        # pick two survivors randomly
        nmate1 = np.random.randint(low=0, high=nkeep+1)
        nmate2 = np.random.randint(low=0, high=nkeep+1)
        
        #then randomly pick DNA from parents for offspring
        
        '''here, do not change property ntemp[nkeep+j+1][0], it's always fixed at the value -1'''
        #if (numpy.random.rand() < 0.5)
        #    ntemp[nkeep+j+1][0] = n[nmate1][0]  
        #else
        #    ntemp[nkeep+j+1][0] = n[nmate2][0]
        # first decide to whether to inherent from nmate1 or nmate2 -> less than 0.5 than inherit from nmate1
        # then decide to how much mutation should be applied
        # then append the offspring at the end: nkeep+j+1
        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][0] = n[nmate1][0]*(1. + mut_c*(0.5 - np.random.rand()))  # property 1, mutation added
        else:
            ntemp[nkeep+j][0] = n[nmate2][0]*(1. + mut_c*(0.5 - np.random.rand()))

        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][1] = n[nmate1][1]*(1. + mut_c*(0.5 - np.random.rand()))  # property 2, mutation added
        else:
            ntemp[nkeep+j][1] = n[nmate2][1]*(1. + mut_c*(0.5 - np.random.rand()))
    
        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][2] = n[nmate1][2]*(1. + mut_c*(0.5 - np.random.rand()))  # property 3, mutation added
        else:
            ntemp[nkeep+j][2] = n[nmate2][2]*(1. + mut_c*(0.5 - np.random.rand()))
    
        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][3] = n[nmate1][3]*(1. + mut_c*(0.5 - np.random.rand()))  # property 4, mutation added
        else:
            ntemp[nkeep+j][3] = n[nmate2][3]*(1. + mut_c*(0.5 - np.random.rand()))

        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][4] = n[nmate1][4]*(1. + mut_c*(0.5 - np.random.rand()))  # property 5, mutation added
        else:
            ntemp[nkeep+j][4] = n[nmate2][4]*(1. + mut_c*(0.5 - np.random.rand()))

    #=============================================
    n = deepcopy(ntemp)   # save ntemp as n for use in next generation (next iteration)
    
    
    '''AVERAGING OVER POPULATION AND OVER TIME, FINDING MINIMUM ERROR SET OF n VALUES'''
    # [6] calculate n1avg[k], etc., which are average n values for population of organisms (solutions) 
    # at this generation k
    #=============================================
    #initialoze average n's to zero and sum contribution of each member of the solution population
     
    n1avg[k] = 0.0
    n2avg[k] = 0.0
    n3avg[k] = 0.0
    n4avg[k] = 0.0
    n5avg[k] = 0.0

    for j in range(NS):
        # Average parameter: divided by total number of gene strand    
        n1avg[k] = n1avg[k] + n[j][0]/NS
        n2avg[k] = n2avg[k] + n[j][1]/NS 
        n3avg[k] = n3avg[k] + n[j][2]/NS
        n4avg[k] = n4avg[k] + n[j][3]/NS
        n5avg[k] = n5avg[k] + n[j][4]/NS

    # Here we compute aFerravgn[i] = absolute Ferr of logrithm data point i using average n values 
    # for this solutions generation k
    # aFerrmeanavgn[k] is the mean of the Ferravgn[i] for the population of organisms/solutions
    # 
    # =============================================
    for i in range(ND): 
        erroravg = 0
        for j in range(int(D*24*60/30)):
            zoneT_NF_avg = np.mean([np.mean(zoneT_in_NF_data_df.iloc[j, 1:-4]), np.mean(zoneT_ex_NF_data_df.iloc[j, 1:-4])])
            zoneT_NS_avg = np.mean([np.mean(zoneT_in_NS_data_df.iloc[j, 1:-4]), np.mean(zoneT_ex_NS_data_df.iloc[j, 1:-4])])
            zoneT_NF_avg_next = np.mean([np.mean(zoneT_in_NF_data_df.iloc[j+1, 1:-4]), np.mean(zoneT_ex_NF_data_df.iloc[j+1, 1:-4])])
            zoneT_NS_avg_next = np.mean([np.mean(zoneT_in_NS_data_df.iloc[j+1, 1:-4]), np.mean(zoneT_ex_NS_data_df.iloc[j+1, 1:-4])])
            outdoorT = np.mean([outdoor_data_df['air_temp_set_1'][j], outdoor_data_df['air_temp_set_2'][j]])
            T_Sf_calc = ((outdoorT - zoneT_NS_avg)/(n1avg[k]*n4avg[k]) + 
                         (zoneT_NF_avg - zoneT_NS_avg)/(n2avg[k]*n4avg[k]) + 
                         (energy_data_df['hvac_N'][j]/2)/n4avg[k] + 
                         (energy_data_df['mels_N'][j]/2)/n4avg[k] +
                         (outdoor_data_df['solar_radiation_set_1'][j]/1000*0.5*A*SHGC*(1 - shade_C)/n4avg[k])*0.5) + zoneT_NS_avg
            T_Ff_calc = ((outdoorT - zoneT_NF_avg)/(n3avg[k]*n5avg[k]) + 
                         (zoneT_NS_avg - zoneT_NF_avg)/(n2avg[k]*n5avg[k]) + 
                         (energy_data_df['hvac_N'][j]/2)/n5avg[k] + 
                         (energy_data_df['mels_N'][j]/2)/n5avg[k] +
                         (outdoor_data_df['solar_radiation_set_1'][j]/1000*0.5*A*SHGC*(1 - shade_C)/n5avg[k])*0.5) + zoneT_NF_avg
            erroravg = erroravg + np.mean([np.abs(T_Ff_calc - zoneT_NF_avg_next), np.abs(T_Sf_calc - zoneT_NS_avg_next)])

        aFerravgn[i] = erroravg/(D*24*60/30)

    #-------------
    aFerrmeanavgn[k] = np.mean(aFerravgn)
    
    # next, update time average of n valaues in population (n1ta[k], etc.) 
    # for generations = k > 800 up to total NGEN
    #=============================================
    aFerrta = aFerrta + aFerrmeanavgn[k]/NGEN
    if (k > NGEN/2):
        n1ta = n1ta + n1avg[k]/(NGEN/2)
        n2ta = n2ta + n2avg[k]/(NGEN/2)
        n3ta = n3ta + n3avg[k]/(NGEN/2)
        n4ta = n4ta + n4avg[k]/(NGEN/2)
        n5ta = n5ta + n5avg[k]/(NGEN/2)
        
    
    # compare aFerrmeanavgn[k] to previous minimum value and save 
    # it and corresponding n(i) values if the value for this generation k is lower
    # If error is lower -> better fit -> update 
    # =============================================
    if (aFerrmeanavgn[k] < aFerrmeanavgnMin):
        aFerrmeanavgnMin = aFerrmeanavgn[k] 
        n1min = n1avg[k]
        n2min = n2avg[k]
        n3min = n3avg[k]
        n4min = n4avg[k]
        n5min = n5avg[k]
    
    #print('avg n1-n4:', n1avg[k], n2avg[k], n3avg[k], n4avg[k], aFerrmeanavgn[k])
    #print ('kvalue =', k)
    '''end of evolution loop'''
    # ------------------------------------------------------------------
    # ------------------------------------------------------------------
    
    
# ------------------------------------------------------------------
#final print and plot of results
# ------------------------------------------------------------------
print('ENDING: pop. avg n1-n5,aFerrmean:', n1avg[k], n2avg[k], n3avg[k], n4avg[k], n5avg[k], aFerrmeanavgn[k])
print('MINUMUM: avg n1-n5,aFerrmeanMin:', n1min, n2min, n3min, n4min, n5min, aFerrmeanavgnMin)
print('TIME AVG: avg n1-n5,aFerrmean:', n1ta, n2ta, n3ta, n4ta, n5ta, aFerrta)

#SETTING UP PLOTS

plt.rcParams.update({'font.size': 18})

# aFerrmeanavgn[k] is the mean of the Ferravgn[i] for the population of organisms/solutions
# computed using the mean n values
plt.plot(gen, aFerrmeanavgn) 
plt.plot(gen, n1avg)
plt.plot(gen, n2avg)
plt.plot(gen, n3avg)
plt.plot(gen, n4avg)
plt.plot(gen, n5avg)
plt.legend(['aFerrmeanavgn', 'n1 avg', 'n2 avg', 'n3 avg', 'n4 avg', 'n5 avg'], loc='lower left')

# naming the x axis 
plt.xlabel('generation') 
# naming the y axis 
plt.ylabel('constants and error') 
plt.loglog()
plt.yticks([0.1,1.0,10])
plt.xticks([1,10,100,1000]) 
plt.show()
```

```python
#- initialize arrays before start of evolution loop EL
#then - create array of DNA strands n[i] and ntemp[i] with dimesnion NS = 5

#i initialize array where rows are dna vectors [n0i,n1i,...n5i] with random pertubations

for i in range(ND):
    n.append([n1i+0.1*random(), 
              n2i+0.1*random(), 
              n3i+0.1*random(), 
              n4i+0.1*random(), 
              n5i+0.1*random(),
              n6i+0.1*random(),
              n7i+0.1*random(),
              n8i+0.1*random(),
              n9i+0.1*random(),
              n10i+0.1*random()])
#print (n) # uncomment command to print array so it can be checked

ntemp = deepcopy(n)
#initialize Ferr values an dother loop parameters
#define arrays of Ferr (error) functions
#individual solution error and absoute error
Ferr = []  
#population average solution error and absoute error
Ferravgn = []
aFerr = []
aFerravgn = []

#store zeros in ND genes
for i in range(ND):
    #individual solution error and absoute error
    Ferr.append([0.0])
    aFerr.append([0.0])
    #population average solution error and absoute error
    Ferravgn.append([0.0])
    aFerravgn.append([0.0])
#print (Ferr)

aFerrmeanavgnMin = 1000000000.0 
# these store the  n values for minimum population average error durng NGEN generations
n1min = 0.0
n2min = 0.0
n3min = 0.0
n4min = 0.0
n5min = 0.0
n6min = 0.0
n7min = 0.0
n8min = 0.0
n9min = 0.0
n10min = 0.0

aFerrta = 0.0
# these store the time averaged n values durng from generation 800 to NGEN generations
n1ta = 0.0
n2ta = 0.0
n3ta = 0.0
n4ta = 0.0
n5ta = 0.0
n6ta = 0.0
n7ta = 0.0
n8ta = 0.0
n9ta = 0.0
n10ta = 0.0
```

```python
'''START OF EVOLUTION LOOP'''
# ------------------------------------------
# k is generation number, NGEN IS TOTAL NUMBER OF GENERATIONS COMPUTED
for k in tqdm(range(NGEN)):
    
    '''In this program , the number of organisms (solutions) NS is taken to be equal to the 
    number of data points ND so for each generation, each solution can be compared to a different 
    data point and all the data is compared in each generation.  The order of the solutions in the array 
    that holds the solution constants is constantly changing due to mating and selection, so the pairing 
    is random.'''
    
    
    '''CALCULATING ERROR (FITNESS)
    In this program, the absolute error in the logrithm of the physical heat flux is 
    used to evaluate fitness.'''
    
    # Here we calculate error Ferr and absolute error aFerr for each data point and organism pair 
    # for specified n(i), and calculate (mean aFerr) = aFerrmean
    # and (median aFerr) = aFerrmedian for the data collection and specified n(i)
    # Note that the number data points ND equals the number of solutions (organisms) in the population
    #====================================================================================
    for i in range(ND):
        error = 0
        for j in range(int(D*24*60/30)):
            zoneT_NF_ex_avg = np.mean(zoneT_ex_NF_data_df.iloc[j, 1:-4])
            zoneT_NF_in_avg = np.mean(zoneT_in_NF_data_df.iloc[j, 1:-4])
            zoneT_NS_ex_avg = np.mean(zoneT_ex_NS_data_df.iloc[j, 1:-4])
            zoneT_NS_in_avg = np.mean(zoneT_in_NS_data_df.iloc[j, 1:-4])
            zoneT_NF_ex_avg_next = np.mean(zoneT_ex_NF_data_df.iloc[j+1, 1:-4])
            zoneT_NF_in_avg_next = np.mean(zoneT_in_NF_data_df.iloc[j+1, 1:-4])
            zoneT_NS_ex_avg_next = np.mean(zoneT_ex_NS_data_df.iloc[j+1, 1:-4])
            zoneT_NS_in_avg_next = np.mean(zoneT_in_NS_data_df.iloc[j+1, 1:-4])
            outdoorT = np.mean([outdoor_data_df['air_temp_set_1'][j], outdoor_data_df['air_temp_set_2'][j]])
            
            T_Sf_ex_calc = ((zoneT_NF_ex_avg - zoneT_NS_ex_avg)/(n[i][2]*n[i][6]) + 
                            (zoneT_NS_in_avg - zoneT_NS_ex_avg)/(n[i][1]*n[i][6]) + 
                            (outdoorT - zoneT_NS_ex_avg)/(n[i][0]*n[i][6]) +
                            (energy_data_df['hvac_N'][j]/4)/(n[i][6]) + 
                            (energy_data_df['mels_N'][j]/4)/(n[i][6]) +
                            (outdoor_data_df['solar_radiation_set_1'][j]/1000*0.5*A*SHGC*(1 - shade_C)/n[i][6])*0.5) + zoneT_NS_ex_avg
            T_Sf_in_calc = ((zoneT_NS_ex_avg - zoneT_NS_in_avg)/(n[i][1]*n[i][7]) + 
                            (zoneT_NF_in_avg - zoneT_NS_in_avg)/(n[i][3]*n[i][7]) + 
                            (energy_data_df['hvac_N'][j]/4)/(n[i][7]) + 
                            (energy_data_df['mels_N'][j]/4)/(n[i][7])) + zoneT_NS_in_avg
            
            T_Ff_ex_calc = ((zoneT_NS_ex_avg - zoneT_NF_ex_avg)/(n[i][2]*n[i][8]) + 
                            (zoneT_NF_in_avg - zoneT_NF_ex_avg)/(n[i][4]*n[i][8]) + 
                            (outdoorT - zoneT_NF_ex_avg)/(n[i][5]*n[i][8]) +
                            (energy_data_df['hvac_N'][j]/4)/(n[i][8]) + 
                            (energy_data_df['mels_N'][j]/4)/(n[i][8]) +
                            (outdoor_data_df['solar_radiation_set_1'][j]/1000*0.5*A*SHGC*(1 - shade_C)/n[i][8])*0.5) + zoneT_NF_ex_avg
            T_Ff_in_calc = ((zoneT_NF_ex_avg - zoneT_NF_in_avg)/(n[i][4]*n[i][9]) + 
                            (zoneT_NS_in_avg - zoneT_NF_in_avg)/(n[i][3]*n[i][9]) + 
                            (energy_data_df['hvac_N'][j]/4)/(n[i][9]) + 
                            (energy_data_df['mels_N'][j]/4)/(n[i][9])) + zoneT_NF_in_avg

            error = error + np.mean([np.abs(T_Ff_ex_calc - zoneT_NF_ex_avg_next),
                                     np.abs(T_Ff_in_calc - zoneT_NF_in_avg_next),
                                     np.abs(T_Sf_ex_calc - zoneT_NS_ex_avg_next),
                                     np.abs(T_Sf_in_calc - zoneT_NS_in_avg_next)])

        aFerr[i] = error/(D*24*60/30) # Assign error for book keeping
    #-------------
    aFerrmean = np.mean(aFerr) # mean error for population for this generation
    aFerrmedian = np.median(aFerr) # median error for population for this generation
    meanAFerr[k]=aFerrmean  #store aFerrmean for this generation gen[k]=k


    '''SELECTION'''
    #pick survivors 
    #[2] calculate survival cutoff, set number kept = nkeep = 0
    #==============================================
    clim = MFRAC*aFerrmedian  #cut off limit is a fraction/multiplier MFRAC of median value for population
    nkeep = 0
    
    # now check each organism/solution to see if aFerr is less than cut of limit
    #if yes, store n for next generation population in ntemp, at end nkeep = number that survive
    #and number of new offspring = NS-nkeep
    #==============================================
    for j in range(NS):  # NS Ferr values, one for each solution in population
        if (aFerr[j] < clim):
            nkeep = nkeep + 1
            ntemp[nkeep-1][0] = n[j][0]
            ntemp[nkeep-1][1] = n[j][1]
            ntemp[nkeep-1][2] = n[j][2]
            ntemp[nkeep-1][3] = n[j][3]
            ntemp[nkeep-1][4] = n[j][4]
            ntemp[nkeep-1][5] = n[j][5]
            ntemp[nkeep-1][6] = n[j][6]
            ntemp[nkeep-1][7] = n[j][7]
            ntemp[nkeep-1][8] = n[j][8]
            ntemp[nkeep-1][9] = n[j][9]

    #now have survivors in leading entries in list of ntemp vectors from 1 to nkeep
    #compute number to be added by mating
    nnew = NS - nkeep  
    
    '''MATING'''
    mut_c = 0.1
    #[4] for nnew new organisms/solutions,
    # randomly pick two survivors, randomly pick DNA (n) from pair for each offspring
    #==============================================================
    for j in range(nnew): 
        # pick two survivors randomly
        nmate1 = np.random.randint(low=0, high=nkeep+1)
        nmate2 = np.random.randint(low=0, high=nkeep+1)
        
        #then randomly pick DNA from parents for offspring
        
        '''here, do not change property ntemp[nkeep+j+1][0], it's always fixed at the value -1'''
        #if (numpy.random.rand() < 0.5)
        #    ntemp[nkeep+j+1][0] = n[nmate1][0]  
        #else
        #    ntemp[nkeep+j+1][0] = n[nmate2][0]
        # first decide to whether to inherent from nmate1 or nmate2 -> less than 0.5 than inherit from nmate1
        # then decide to how much mutation should be applied
        # then append the offspring at the end: nkeep+j+1
        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][0] = n[nmate1][0]*(1. + mut_c*(0.5 - np.random.rand()))  # property 1, mutation added
        else:
            ntemp[nkeep+j][0] = n[nmate2][0]*(1. + mut_c*(0.5 - np.random.rand()))

        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][1] = n[nmate1][1]*(1. + mut_c*(0.5 - np.random.rand()))  # property 2, mutation added
        else:
            ntemp[nkeep+j][1] = n[nmate2][1]*(1. + mut_c*(0.5 - np.random.rand()))
    
        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][2] = n[nmate1][2]*(1. + mut_c*(0.5 - np.random.rand()))  # property 3, mutation added
        else:
            ntemp[nkeep+j][2] = n[nmate2][2]*(1. + mut_c*(0.5 - np.random.rand()))
    
        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][3] = n[nmate1][3]*(1. + mut_c*(0.5 - np.random.rand()))  # property 4, mutation added
        else:
            ntemp[nkeep+j][3] = n[nmate2][3]*(1. + mut_c*(0.5 - np.random.rand()))

        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][4] = n[nmate1][4]*(1. + mut_c*(0.5 - np.random.rand()))  # property 5, mutation added
        else:
            ntemp[nkeep+j][4] = n[nmate2][4]*(1. + mut_c*(0.5 - np.random.rand()))
        
        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][5] = n[nmate1][5]*(1. + mut_c*(0.5 - np.random.rand()))  # property 5, mutation added
        else:
            ntemp[nkeep+j][5] = n[nmate2][5]*(1. + mut_c*(0.5 - np.random.rand()))

        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][6] = n[nmate1][6]*(1. + mut_c*(0.5 - np.random.rand()))  # property 5, mutation added
        else:
            ntemp[nkeep+j][6] = n[nmate2][6]*(1. + mut_c*(0.5 - np.random.rand()))

        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][7] = n[nmate1][7]*(1. + mut_c*(0.5 - np.random.rand()))  # property 5, mutation added
        else:
            ntemp[nkeep+j][7] = n[nmate2][7]*(1. + mut_c*(0.5 - np.random.rand()))

        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][8] = n[nmate1][8]*(1. + mut_c*(0.5 - np.random.rand()))  # property 5, mutation added
        else:
            ntemp[nkeep+j][8] = n[nmate2][8]*(1. + mut_c*(0.5 - np.random.rand()))
        
        if (np.random.rand() < 0.5):
            ntemp[nkeep+j][9] = n[nmate1][9]*(1. + mut_c*(0.5 - np.random.rand()))  # property 5, mutation added
        else:
            ntemp[nkeep+j][9] = n[nmate2][9]*(1. + mut_c*(0.5 - np.random.rand()))

    #=============================================
    n = deepcopy(ntemp)   # save ntemp as n for use in next generation (next iteration)
    
    
    '''AVERAGING OVER POPULATION AND OVER TIME, FINDING MINIMUM ERROR SET OF n VALUES'''
    # [6] calculate n1avg[k], etc., which are average n values for population of organisms (solutions) 
    # at this generation k
    #=============================================
    #initialoze average n's to zero and sum contribution of each member of the solution population
     
    n1avg[k] = 0.0
    n2avg[k] = 0.0
    n3avg[k] = 0.0
    n4avg[k] = 0.0
    n5avg[k] = 0.0
    n6avg[k] = 0.0
    n7avg[k] = 0.0
    n8avg[k] = 0.0
    n9avg[k] = 0.0
    n10avg[k] = 0.0

    for j in range(NS):
        # Average parameter: divided by total number of gene strand    
        n1avg[k] = n1avg[k] + n[j][0]/NS
        n2avg[k] = n2avg[k] + n[j][1]/NS 
        n3avg[k] = n3avg[k] + n[j][2]/NS
        n4avg[k] = n4avg[k] + n[j][3]/NS
        n5avg[k] = n5avg[k] + n[j][4]/NS
        n6avg[k] = n6avg[k] + n[j][5]/NS
        n7avg[k] = n7avg[k] + n[j][6]/NS
        n8avg[k] = n8avg[k] + n[j][7]/NS
        n9avg[k] = n9avg[k] + n[j][8]/NS
        n10avg[k] = n10avg[k] + n[j][9]/NS

    # Here we compute aFerravgn[i] = absolute Ferr of logrithm data point i using average n values 
    # for this solutions generation k
    # aFerrmeanavgn[k] is the mean of the Ferravgn[i] for the population of organisms/solutions
    # 
    # =============================================
    for i in range(ND): 
        erroravg = 0
        for j in range(int(D*24*60/30)):
            zoneT_NF_ex_avg = np.mean(zoneT_ex_NF_data_df.iloc[j, 1:-4])
            zoneT_NF_in_avg = np.mean(zoneT_in_NF_data_df.iloc[j, 1:-4])
            zoneT_NS_ex_avg = np.mean(zoneT_ex_NS_data_df.iloc[j, 1:-4])
            zoneT_NS_in_avg = np.mean(zoneT_in_NS_data_df.iloc[j, 1:-4])
            zoneT_NF_ex_avg_next = np.mean(zoneT_ex_NF_data_df.iloc[j+1, 1:-4])
            zoneT_NF_in_avg_next = np.mean(zoneT_in_NF_data_df.iloc[j+1, 1:-4])
            zoneT_NS_ex_avg_next = np.mean(zoneT_ex_NS_data_df.iloc[j+1, 1:-4])
            zoneT_NS_in_avg_next = np.mean(zoneT_in_NS_data_df.iloc[j+1, 1:-4])
            outdoorT = np.mean([outdoor_data_df['air_temp_set_1'][j], outdoor_data_df['air_temp_set_2'][j]])
            
            T_Sf_ex_calc = ((zoneT_NF_ex_avg - zoneT_NS_ex_avg)/(n3avg[k]*n7avg[k]) + 
                            (zoneT_NS_in_avg - zoneT_NS_ex_avg)/(n2avg[k]*n7avg[k]) + 
                            (outdoorT - zoneT_NS_ex_avg)/(n1avg[k]*n7avg[k]) +
                            (energy_data_df['hvac_N'][j]/4)/(n7avg[k]) + 
                            (energy_data_df['mels_N'][j]/4)/(n7avg[k]) +
                            (outdoor_data_df['solar_radiation_set_1'][j]/1000*0.5*A*SHGC*(1 - shade_C)/n7avg[k])*0.5) + zoneT_NS_ex_avg
            T_Sf_in_calc = ((zoneT_NS_ex_avg - zoneT_NS_in_avg)/(n2avg[k]*n8avg[k]) + 
                            (zoneT_NF_in_avg - zoneT_NS_in_avg)/(n4avg[k]*n8avg[k]) + 
                            (energy_data_df['hvac_N'][j]/4)/(n8avg[k]) + 
                            (energy_data_df['mels_N'][j]/4)/(n8avg[k])) + zoneT_NS_in_avg
            
            T_Ff_ex_calc = ((zoneT_NS_ex_avg - zoneT_NF_ex_avg)/(n3avg[k]*n9avg[k]) + 
                            (zoneT_NF_in_avg - zoneT_NF_ex_avg)/(n5avg[k]*n9avg[k]) + 
                            (outdoorT - zoneT_NF_ex_avg)/(n6avg[k]*n9avg[k]) +
                            (energy_data_df['hvac_N'][j]/4)/(n9avg[k]) + 
                            (energy_data_df['mels_N'][j]/4)/(n9avg[k]) +
                            (outdoor_data_df['solar_radiation_set_1'][j]/1000*0.5*A*SHGC*(1 - shade_C)/n9avg[k])*0.5) + zoneT_NF_ex_avg
            T_Ff_in_calc = ((zoneT_NF_ex_avg - zoneT_NF_in_avg)/(n5avg[k]*n10avg[k]) + 
                            (zoneT_NS_in_avg - zoneT_NF_in_avg)/(n4avg[k]*n10avg[k]) + 
                            (energy_data_df['hvac_N'][j]/4)/(n10avg[k]) + 
                            (energy_data_df['mels_N'][j]/4)/(n10avg[k])) + zoneT_NF_in_avg

            erroravg = erroravg + np.mean([np.abs(T_Ff_ex_calc - zoneT_NF_ex_avg_next),
                                           np.abs(T_Ff_in_calc - zoneT_NF_in_avg_next),
                                           np.abs(T_Sf_ex_calc - zoneT_NS_ex_avg_next),
                                           np.abs(T_Sf_in_calc - zoneT_NS_in_avg_next)])

        aFerravgn[i] = erroravg/(D*24*60/30)

    #-------------
    aFerrmeanavgn[k] = np.mean(aFerravgn)
    
    # next, update time average of n valaues in population (n1ta[k], etc.) 
    # for generations = k > 800 up to total NGEN
    #=============================================
    aFerrta = aFerrta + aFerrmeanavgn[k]/NGEN
    if (k > NGEN/2):
        n1ta = n1ta + n1avg[k]/(NGEN/2)
        n2ta = n2ta + n2avg[k]/(NGEN/2)
        n3ta = n3ta + n3avg[k]/(NGEN/2)
        n4ta = n4ta + n4avg[k]/(NGEN/2)
        n5ta = n5ta + n5avg[k]/(NGEN/2)
        n6ta = n6ta + n6avg[k]/(NGEN/2)
        n7ta = n7ta + n7avg[k]/(NGEN/2)
        n8ta = n8ta + n8avg[k]/(NGEN/2)
        n9ta = n9ta + n9avg[k]/(NGEN/2)
        n10ta = n10ta + n10avg[k]/(NGEN/2)
        
    
    # compare aFerrmeanavgn[k] to previous minimum value and save 
    # it and corresponding n(i) values if the value for this generation k is lower
    # If error is lower -> better fit -> update 
    # =============================================
    if (aFerrmeanavgn[k] < aFerrmeanavgnMin):
        aFerrmeanavgnMin = aFerrmeanavgn[k] 
        n1min = n1avg[k]
        n2min = n2avg[k]
        n3min = n3avg[k]
        n4min = n4avg[k]
        n5min = n5avg[k]
        n6min = n6avg[k]
        n7min = n7avg[k]
        n8min = n8avg[k]
        n9min = n9avg[k]
        n10min = n10avg[k]
    
    #print('avg n1-n4:', n1avg[k], n2avg[k], n3avg[k], n4avg[k], aFerrmeanavgn[k])
    #print ('kvalue =', k)
    '''end of evolution loop'''
    # ------------------------------------------------------------------
    # ------------------------------------------------------------------
    
    
# ------------------------------------------------------------------
#final print and plot of results
# ------------------------------------------------------------------
print('ENDING: pop. avg n1-n5,aFerrmean:', n1avg[k], 
                                           n2avg[k], 
                                           n3avg[k], 
                                           n4avg[k], 
                                           n5avg[k],
                                           n6avg[k],
                                           n7avg[k],
                                           n8avg[k],
                                           n9avg[k],
                                           n10avg[k],
                                           aFerrmeanavgn[k])
print('MINUMUM: avg n1-n5,aFerrmeanMin:', n1min, 
                                          n2min, 
                                          n3min, 
                                          n4min, 
                                          n5min, 
                                          n6min,
                                          n7min,
                                          n8min,
                                          n9min,
                                          n10min,
                                          aFerrmeanavgnMin)
print('TIME AVG: avg n1-n5,aFerrmean:', n1ta, 
                                        n2ta, 
                                        n3ta, 
                                        n4ta, 
                                        n5ta, 
                                        n6ta,
                                        n7ta,
                                        n8ta,
                                        n9ta,
                                        n10ta,
                                        aFerrta)

#SETTING UP PLOTS

plt.rcParams.update({'font.size': 18})

# aFerrmeanavgn[k] is the mean of the Ferravgn[i] for the population of organisms/solutions
# computed using the mean n values
plt.plot(gen, aFerrmeanavgn) 
plt.plot(gen, n1avg)
plt.plot(gen, n2avg)
plt.plot(gen, n3avg)
plt.plot(gen, n4avg)
plt.plot(gen, n5avg)
plt.plot(gen, n6avg)
plt.plot(gen, n7avg)
plt.plot(gen, n8avg)
plt.plot(gen, n9avg)
plt.plot(gen, n10avg)
plt.legend(['aFerrmeanavgn', 
            'n1 avg', 
            'n2 avg', 
            'n3 avg', 
            'n4 avg', 
            'n5 avg',
            'n6 avg',
            'n7 avg',
            'n8 avg',
            'n9 avg',
            'n10 avg',], loc='lower left')

# naming the x axis 
plt.xlabel('generation') 
# naming the y axis 
plt.ylabel('constants and error') 
plt.loglog()
plt.yticks([0.1,1.0,10])
plt.xticks([1,10,100,1000]) 
plt.show()
```

```python
def RC(RC_list, outdoorT, zoneT_list, u, s, m):
    """
    RC_list: R and C values
    outdoorT: outdoor temperature
    zoneT_list: zone temperature (0: first floor; 1: second floor)
    u: power input
    s: solar radiation
    m: other power input (Miscellaneous electric power)
    """
    T_2 = ((outdoorT - zoneT_list[1])/(RC_list[0]*RC_list[3]) + 
           (zoneT_list[0] - zoneT_list[1])/(RC_list[1]*RC_list[3]) + 
           (u)/RC_list[3] +
           (m)/RC_list[3] +
           (s/1000*0.5*A*SHGC*(1 - shade_C)/RC_list[3])*0.5) + zoneT_list[1]

    T_1 = ((outdoorT - zoneT_list[0])/(RC_list[2]*RC_list[4]) + 
           (zoneT_list[1] - zoneT_list[0])/(RC_list[1]*RC_list[4]) + 
           (u)/RC_list[4] +
           (m)/RC_list[4] +
           (s/1000*0.5*A*SHGC*(1 - shade_C)/RC_list[4])*0.5) + zoneT_list[0]
    
    T = list([T_1, T_2])
    return T
```
